```python
import pandas as pd
import geodatasets
import geopandas as gpd
from shapely.geometry import Point
import matplotlib.pyplot as plt
import folium
```

```python
# Load the data from the file
df = pd.read_csv('Output.csv')

# Drop rows where 'lat_long_manual' is NA
df = df.dropna(subset=['lat_long_manual'])

# world = gpd.read_file(geodatasets.get_path('naturalearth_cultural'))
```

```python
# Split the 'lat_long_manual' column into 'latitude' and 'longitude'
df[['lat', 'long']] = df['lat_long_manual'].str.split(', ', expand=True)
df['lat'] = pd.to_numeric(df['lat'])
df['long'] = pd.to_numeric(df['long'])

# Create a base map centered around the average of the coordinates
m = folium.Map(location=[df['lat'].mean(), df['long'].mean()], zoom_start=2)

# Add points to the map
for _, row in df.iterrows():
    folium.Marker(location=[row['lat'], row['long']]).add_to(m)

# Save the map to an HTML file
m.save('map.html')

# If you want to display it directly in a Jupyter Notebook, use:
m
```
